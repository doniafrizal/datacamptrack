```python
# Import library for file handling
import os
```

```python
# Import library for data processing
import pandas as pd
```

```python
# Import library for plotting
from matplotlib import pyplot as plt


# %%

# Import air quality file as pandas DataFrame
air_quality = pd.read_csv(os.path.join('00._data', 'air-quality.csv'), parse_dates=['Date'], index_col='Date')
```

```python
# Display head
print(air_quality.head())
```

```python
# Number of nullity
air_quality.isnull().sum()
```

```python
# Percentage of nullity
air_quality.isnull().mean() * 100
```

```python
# Forward Fill Imputation
air_quality_ffill = air_quality.copy(deep=True)

# Check nullity
print('Nullity before:\n', air_quality_ffill['Ozone'][30:40])

# Start Forward Fill Imputation
air_quality_ffill.fillna(method='ffill', inplace=True)

# Verify Nullity
print('\n Nullity after:\n', air_quality_ffill['Ozone'][30:40])
```

```python
# Backward Fill Imputation
air_quality_bfill = air_quality.copy(deep=True)

# Check nullity
print('Nullity before:\n', air_quality_bfill['Ozone'][30:40])

# Start Backward Fill Imputation
air_quality_bfill.fillna(method='bfill', inplace=True)

# Verify Nullity
print('\n Nullity after:\n', air_quality_bfill['Ozone'][30:40])
```

```python
# Interpolate linear Fill Imputation
air_quality_ilinear = air_quality.copy(deep=True)

# Check nullity
print('Nullity before:\n', air_quality_ilinear['Ozone'][30:40])

# Start Interpolate Linear Imputation
air_quality_ilinear.interpolate(method='linear', inplace=True)

# Verify Nullity
print('\n Nullity after:\n', air_quality_ilinear['Ozone'][30:40])
```

```python
# Interpolate linear Fill Imputation
air_quality_iquad = air_quality.copy(deep=True)

# Check nullity
print('Nullity before:\n', air_quality_iquad['Ozone'][30:40])

# Start Interpolate Linear Imputation
air_quality_iquad.interpolate(method='quadratic', inplace=True)

# Verify Nullity
print('\n Nullity after:\n', air_quality_iquad['Ozone'][30:40])
```

```python
# Interpolate linear Fill Imputation
air_quality_inear = air_quality.copy(deep=True)

# Check nullity
print('Nullity before:\n', air_quality_inear['Ozone'][30:40])

# Start Interpolate Linear Imputation
air_quality_inear.interpolate(method='nearest', inplace=True)

# Verify Nullity
print('\n Nullity after:\n', air_quality_inear['Ozone'][30:40])
```

```python
# Visualizing Imputations
# Set nrows to 3 and ncols to 1
fig, axes = plt.subplots(3, 1, figsize=(30, 20))

# Create a dictionary of interpolations
interpolations = {'Linear Interpolation': air_quality_ilinear,
                  'Quadratic Interpolation': air_quality_iquad,
                  'Nearest Interpolation': air_quality_inear}

# Loop over axes and interpolations
for ax, df_key in zip(axes, interpolations):
  # Select and also set the title for a DataFrame
  interpolations[df_key].Ozone.plot(color='red', marker='o',
                                 linestyle='dotted', ax=ax)
  air_quality.Ozone.plot(title=df_key + ' - Ozone', marker='o', ax=ax)

plt.show()
```
